```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
df_raw: tuple = load_iris(as_frame=True, return_X_y=True)
df_main: pd.DataFrame = df_raw[0]
df_target: pd.DataFrame = pd.merge(df_raw[0], df_raw[-1], how='inner', left_index=True, right_index=True)
df_target
```

```python
plt.style.use('default')
sns.set_theme(rc={'figure.dpi': 300})
fig = sns.pairplot(data=df_target, hue='target')
fig
```
